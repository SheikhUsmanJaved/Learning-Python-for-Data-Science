# EDA (Exploratory data analysis)
EDA is a method used to `Analyze` and `Summarize` datasets. In python, you can use the pandas library to perform EDA on a dataset.
EDA is a crucial step in the data science process. It involves using statistical and visual methods to understand


```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# We need to analyze and summarize the data

```python
df = sns.load_dataset('titanic')
```

```python
# Finding no. of unique values
#  No. of unique values required to explore the variables that how many unique values it contains.
# If a variable contains more than 10 unique values then it 90% of chances are that it contains a numeric value. Sometimes even if the unique value 
# crosses a no. of 20 it is a categorical variable.
df.nunique()
```

```python
df.columns
```

```python
# Visulaize the nan values in heatmap
sns.heatmap(df.isnull())
```

```python
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
# cmap='viridis'is a color pallette (check & try what is this)
# cbar=False is color bar (check & try what is this)
```

```python
df.drop(columns='deck', inplace=True)
```

```python
df.columns
```

```python
df.shape
```

```python
# Visualize and explore
# countplot doesn;t need y axis value. On y axis it automatically takes counting number 
sns.countplot(data=df, x='survived')
```

```python
sns.barplot(data=df, x="survived", y="age", hue="who")
```

```python
# x-axis categirocal variable and on y-axis numerical variable 
sns.barplot(data=df, x="survived", y="fare", hue="who")
```

```python
# Only categorical variables
sns.countplot(x="pclass", data=df)
```

```python
# Data distribution
sns.histplot(data=df, x="age")
```

```python
#  Creating bins means grouping data. 
sns.histplot(data=df, x="age", bins=5)
```

```python
# Hist plot is possible with numerical variable 
df['age'].hist(bins=30, color='darkred', alpha=0.7)
# alpha=0.7 means opacity 
```

```python
sns.boxplot(data=df, x="age")
```

```python
# Is it possible to create a boxplot for two numerical variables 
# It is possible if one variable is categorical na dthe second is numerical
sns.boxplot(data=df, y='age', x='survived')
```

```python
sns.boxplot(data=df, y='age', x='survived', hue='who')
```

```python
# scatter plot is possible with both numerical variables 
sns.scatterplot(x="age", y="fare", data=df)
```

```python
sns.scatterplot(x="age", y="fare", data=df, hue="pclass")
plt.ylim(0,250)
```

```python
sns.pairplot(df, hue="survived")
```

```python
#  Stacked barchart/plot
df.groupby(['pclass', 'sex'])['survived'].mean().unstack().plot(kind='bar', stacked=True)
```

```python
df.groupby(['pclass', 'who'])['survived'].mean().unstack().plot(kind='bar', stacked=True)
```

```python
df.groupby(['pclass', 'who'])['fare'].mean().unstack().plot(kind='bar', stacked=True)
```

```python
# Swarm Plot 
# It is similar to scatter plot but if you plot swarm plot on bar plot it will show the data points 
sns.swarmplot(data=df, x='age', y='fare')
```

```python
sns.swarmplot(data=df, x='survived', y='fare')
```

```python
sns.swarmplot(data=df, x='survived', y='age')
```

```python
sns.swarmplot(data=df, x='survived', y='age', hue='who')
```

```python
# Catplot / point plot
sns.catplot(x='pclass', y='fare', hue='embarked', kind='point', data=df)
```

```python
# LM plot (relation of two numeric variables)
sns.lmplot(data=df, x='age', y='fare')
```

```python
sns.lmplot(data=df[df['fare']<300], x='age', y='fare', hue='pclass')
```

```python
# Assignment: use conditional argument on y-axis (fare) to remove the outliar and to draw the graph at 300.
sns.lmplot(data=df, x='age', y='fare', height=4)
```

```python
# How to add a column and make an age group
sns.histplot(data=df, x='age')
```

```python
# Grouping
df.columns
```

```python
df['age_group'] = pd.cut(df['age'], [0,10,18,30,50,80],
                         labels=['kids', 'teen_agers', 'youth', 'middle_age', 'old'])
```

```python
df.head()
```

```python
df.groupby('age_group')['survived'].mean().plot(kind="bar")
```

```python
df.groupby('age_group')['fare'].mean().plot(kind="bar")
```

```python
sns.pairplot(df, hue='survived', vars=['age', 'fare', 'pclass'])
```
