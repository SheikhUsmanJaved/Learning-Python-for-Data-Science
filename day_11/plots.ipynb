```python
# import libraries

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import dataset
kashti = sns.load_dataset('titanic')
kashti
```

```python
# check konsa parameter kia bata raha hai
kashti.info()
```

```python
# Before plotting check the null values

kashti.isnull().sum()
```

```python
# Example of plot
kashti.survived.value_counts().plot(kind="bar")
```

```python
# Another Example
sns.countplot(x= "survived", hue="sex", data=kashti)
```

```python
# Scatter plot example
plt.scatter(data=kashti, x="age", y="fare")

# Assignment: How to change color (hue) using plt library
```

```python
# above plot using sns

sns.scatterplot(data=kashti, x="age", y="fare", hue="pclass")
```

```python
# above example with sns.barplot

sns.barplot(data=kashti, x="sex", y="fare", hue="pclass")
```

```python
# An important plot
# Pairplot means to draw a plot of each colums in a pair with other columns 
sns.pairplot(kashti)
```

```python
# Histogram of all the numeric variables.
# bars are called bins


kashti.hist(bins=10, figsize=(5,7), grid=False)
```

```python
# Box plot
sns.boxplot(data=kashti, x="sex", y="fare", hue="pclass")
```

```python
# Boxenplot
sns.boxenplot(data=kashti, x="sex", y="fare", hue="pclass")
```

```python
#Assignment: draw new graphs by changing the x-axis and y-axis titles.
```

```python
# Facet Grid plot

g= sns.FacetGrid(kashti, hue = "survived", col = "pclass", margin_titles = True,
                 palette= {1:"seagreen", 0:"gray"})
g = g.map(plt.scatter, "fare", "age", edgecolor="w").add_legend();
```

```python
# Strip plot (mix of bar and scatter plot)
sns.stripplot(data=kashti, x="sex", y="fare", hue="pclass")
```

```python
# Heat map
# corr=corelation
corr = kashti.corr()
# plt.figure(figsize=(10,10))
# sns.heatmap(corr)
#plt.title("heat map of titanic")
```

```python
# BViolinplot
sns.violinplot(data=kashti, x="sex", y="fare", hue="class")
```
