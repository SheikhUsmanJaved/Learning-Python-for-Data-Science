# 16- Splitting a string into multiple columns

```python

#Splitting a data set columns is very important, Mostly people working with time series faces issues and needs to use this tip.

#Defining a dataset
df = pd.DataFrame({'name' : ['Abubakar Tariq', 'Umar Saleem', 'Usman Javed', 'Muhammad Ali'],
                   'Location' : ['Helsinki, Germany', 'Beijing, China', 'Lahore, Pakistan', 'Riyadh, SaudiArabia']})
df
```

```python
#How to split a column values

df.name.str.split(' ').head()
```

```python
#To split the values in separate columns
df.name.str.split(' ', expand=True)
```

```python
#To add the splitted values in new colums at the end of the same dataset 
df[['first name', 'last name']] = df.name.str.split(' ', expand=True)
df
```

```python
#Assignment: expand/split location values
df.Location.str.split(' ', expand=True)
```

```python
df[['City', 'Country']] = df.Location.str.split(' ', expand=True)
df
```

```python
# Refined data manipulation / to use the new splitted data only

df = df[['first name', 'last name', 'City', 'Country']]
df
```

# 17- Aggregate by multiple groups / functions

```python
#import libraries
import pandas as pd
import seaborn as sns

#import dataset
df = sns.load_dataset('titanic')
df.head()
```

```python
# Now we will check how we can group the data within the same dataset using different functions
df.groupby('who').count()
```

```python
df.groupby('sex').count()
```

```python
df.groupby('sex').head()
```

```python
df.dtypes
```

```python
#using sum function

numeric_columns = df.select_dtypes(include=('number', 'bool')).columns
df.groupby('who')[numeric_columns].sum()
```

```python
#length is three because the category "who" contains 3 values/categories (child, man, woman)
len(df.groupby('who'))
```

```python
#length is 2 because the category "sex" contains 2 values/categories (male, female)
len(df.groupby('sex'))
```

```python
# To group up based on two or more categories

df.groupby(['sex', 'who', 'pclass']).count()
```
